```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import json
import re
import os
import glob
import PyPDF2
from PyPDF2 import PdfFileReader
from tika import parser
from collections import Counter
```

```python
# Step 1: Read in the data
akababi_science_grade_3 = parser.from_file('Akababi Science Grade 3.pdf')

# Step 2: Extract the text from the pdf
akababi_science_grade_3_text = akababi_science_grade_3['content']
print(akababi_science_grade_3_text)
```

```python
# We are particularly interested in parsing the text by chapter/topic ie the ones in bold
# Step 3: Extract the chapters/topics
# We will use regex to extract the chapters/topics
# The chapters/topics are in bold and are followed by a number
# We will use the regex pattern to extract the chapters/topics
# The regex pattern is as follows:
# 1. The chapters/topics are in bold and are preceded by a number

# This is a regex pattern that matches a number followed by any text (which represents the chapter/topic title).
# \d+ matches one or more digits
# \.\s matches a dot followed by a space
# .* matches any character (except a newline) between zero and unlimited times
chapter_pattern = r'\d+\.\s.*'

# This splits the text into a list of chapters, but it removes the chapter headings.
chapters = re.split(chapter_pattern, akababi_science_grade_3_text)

# If you want to keep the chapter headings in the text, you can use parentheses in the pattern:
chapters = re.split("(" + chapter_pattern + ")", akababi_science_grade_3_text)

# This will result in a list where each chapter heading is followed by the text of the chapter.
print(chapters)
```

```python
# Assuming 'chapters' is a list of strings where each string is a chapter
cleaned_chapters = [chapter.replace('\n', ' ') for chapter in chapters]

# Now you can write the cleaned chapters to a CSV or JSON file
```

```python
print(cleaned_chapters)
```

```python
import csv

# Assuming 'chapters' is a list of strings where each string is a chapter
with open('chapters.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    for chapter in cleaned_chapters:
        writer.writerow([chapter])
```

```python
import json

# Assuming 'chapters' is a list of strings where each string is a chapter
with open('chapters.json', 'w', encoding='utf-8') as file:
    json.dump(cleaned_chapters, file, ensure_ascii=False, indent=4)
```
