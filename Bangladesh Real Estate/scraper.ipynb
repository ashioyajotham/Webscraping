```python
ample_property_url = "https://propertyconnectbd.com/property/ample-apartment/"

# headers
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:102.0) Gecko/20100101 Firefox/102.0"
}

# Send a GET request to the URL
response = requests.get(ample_property_url, headers=headers)

# Print the status code
print(response.status_code)
```

```python
# Extracting info using BS
soup = BeautifulSoup(response.content)
print(soup.prettify)
```

```python
# As we can see h2 contains most useful info so
# we will scrape from there

# Get the text
soup.find_all("h2")[1].text
```

```python
# Scrape the details in a list
details = []
for i in range(1, 5):
    details.append(soup.find_all("h2")[i].text)

# Print the details
print(details)
```
